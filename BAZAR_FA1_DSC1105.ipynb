```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import scipy.stats as stats


df = pd.read_csv("cytof_one_experiment.csv")

column = "GranzymeB"
data = df[column]

fig, axes = plt.subplots(1, 2, figsize=(12, 5))


sns.histplot(data, bins=50, kde=True, ax=axes[0], color="blue")
axes[0].set_title(f"Histogram of {column}")
axes[0].set_xlabel(column)
axes[0].set_ylabel("Frequency")


sorted_data = np.sort(data)
ecdf = np.arange(1, len(sorted_data) + 1) / len(sorted_data)
axes[1].plot(sorted_data, ecdf, marker=".", linestyle="none", color="red")
axes[1].set_title(f"ECDF of {column}")
axes[1].set_xlabel(column)
axes[1].set_ylabel("ECDF")

plt.tight_layout()
plt.show()
```

- The shape of the distribution appears right-skewed (positively skewed), there are more values concentrated on the left with a tail extending towards higher values. There is also a noticeable peak around 3.54.0, suggesting that most values are within this range. 

- The ECDF shows the cumulative proportion of data points.
A steep rise in the ECDF curve between 3.0 and 4.5 suggests that a large proportion of data falls in this range. The curve flattens at the right end, indicating fewer data points in the upper range.

```python
column1 = "GranzymeB"
column2 = "Perforin"


fig, ax = plt.subplots(figsize=(6, 6))
stats.probplot(df[column1], dist="norm", plot=ax)
stats.probplot(df[column2], dist="norm", plot=ax)

ax.set_title(f"Q-Q Plot: {column1} vs {column2}")
ax.legend([column1, column2])
plt.show()
```

- This Q-Q plot suggests that the GranzymeB (blue points) are more spread out, it has higher variability and heavier tails, and potentially more extreme values. The Perforin shows a different curvature, indicating that it may have less skewness compared to GranzymeB.
